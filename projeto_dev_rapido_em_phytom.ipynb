```python
#começando a puxar dados de mortes por ano no continente africano e somando para assim conseguir colocar no grafico

dados_por_ano_africa = africa_df.groupby('Year')
#usando esse de baixo pra conerir se agrupou coretamente 
#dados_1990 = dados_por_ano_africa.get_group(1990)
#somando os dados dos paises da africa 
soma_por_ano_africa = dados_por_ano_africa.sum()
#chamando esses dados para ver como ficou
soma_por_ano_africa
```

```python
#plotando o primeiro grafico com relaçao a mortes no continente africano de crianças menores de 5 anos a cada ano
import plotly.graph_objs as go
import plotly.offline as py
import plotly.graph_objs as go

data = [go.Bar(x=africa_df['Year'].values,
               y=soma_por_ano_africa['Deaths - HIV/AIDS - Sex: Both - Age: Under 5 (Number)']
               )
        ]
               

#criando layout
configuracoes_layout = go.Layout(title='mortes no continente africano a cada ano de pessoas com menos de 5 anos',
                                 yaxis={'title': 'numero de mortes'},
                                 xaxis={'title': 'Período'})

#objeto figura

fig = go.Figure(data=data, layout=configuracoes_layout)

#plotando o grafico
py.iplot(fig)
#pyo.plot(fig)
```
